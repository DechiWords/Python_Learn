## Números Complejos

Podemos definir los números complejos como pares ordenados $(x,y)$ de números reales, $\mathbb R$, interpretados como puntos del plano complejo con coordenadas rectangulares $x$ e $y$.

Denotamos el número complejo $(x,y)$ por $z$. Los números reales $x$ e $y$ en este caso se llaman la parte real y la parte imaginaria de $z$.

Todo número complejo $z = (x,y)$ puede escribirse como $z = (x,0) + (0,y)$ y es fácil verificar que $(0,1)(y,0) = (0,y)$. Se sigue entonces que

$$z = (x,0) + (0,1)(y,0)$$

Así, si pensamos en los números reales como $x$ o $(x,0)$, indistintivamente, y si denotamos por $i$ el imaginario puro $(0,1)$, podemos escribir

$$z = x + iy$$

Python acepta los números complejos y de hecho están bien definidos en este lenguaje, sin embargo, a diferencia de su notación matemática, $a+bi$, Python identifica a los números complejos de la siguiente forma $a+bj$ donde $R(z) = a, Im(z) = bj, z\in \mathbb C$.

````python
type(2+3j)
````

Podemos definir un número complejo, computacionalmente hablando, mediante la función **complex**

````python
complex(2)
````

Es posible obtener el regreso de la parte imaginaria o real de un número complejo mediante la instrucción **real** e **imag**

````python
(9+9j).real
````

````python
(9+45j).imag
````

Si bien sabemos que $\mathbb R\subset \mathbb C$ para una computadora existe una diferencia incluso entre los tipos de datos.

La librería **cmath** contiene funciones especiales para realizar trabajos especiales con números complejos.

Recordemos ciertos aspectos de los números complejos.

### Suma de números complejos

Sean $z_1,z_2\in \mathbb C,z_1 = (x_1,y_1),z_2=(x_2,y_2)$, definimos la suma de $z_1+z_2$ de la siguiente manera

$$z_1 + z_2 = (x_1,y_1) + (x_2,y_2) = (x_1+x_2,y_1+y_2) = (x_1 + x_2) + i(y_1 + y_2)$$

En Python esta operación esta bien definida

````python
(8+9j) + (-8+9j)
````

### Producto de números complejos

Sean $z_1,z_2\in \mathbb C,z_1 = (x_1,y_1),z_2=(x_2,y_2)$, definimos la suma de $z_1\cdot z_2$ de la siguiente manera

$$z_1\cdot z_2 = (x_1,y_1)\cdot (x_2,y_2) = (x_1x_2-y_1y_2,y_1x_2+x_1y_2) = (x_1x_2-y_1y_2)+i(y_1x_2 + x_1y_2)$$

En Python esta operación esta bien definida

````python
(2+2j)*(4+4j)
````

### $\mathbb C$

El conjunto $\mathbb C$ es un anillo conmutativo, esto es, que bajo la suma y el producto definidos anteriormente tenemos que se cumple 

* Cerradura

* Conmutatividad

* Asociatividad

* Elemento Neutro (Unitario para el producto)

* Elemento Inverso

* Leyes distributivas ($z(z_1+z_2) = zz_1+zz_2$)

Tomemos en cuenta que el elemento neutro para el producto y la suma son, respectivamente, $(1,0),(0,0)$.

````python
1*(2+2j)
````

````python
(4+4j)+0
````

El elemento inverso de un número complejo no cero para el producto es denotado como $z^{-1}$ y está dado como

$$z^{-1} = \left(\frac{x}{x^2+y^2},\frac{-y}{x^2+y^2}\right), z\neq 0$$

````python
(4+1j)*( (4/(16+1)) -(1j)/(16+1) ) #z*z^-1
````

En Python la división de números complejos esta bien definida

````python
(0+1j)/(1+2j)
````

### Módulos

Los números complejos también pueden ser representados por segmentos de recta dirigidos o vectores. Bajo ésta interpretación, la suma de números complejos también es la suma como vectores, sin embargo **la multiplicación no tiene análogo**

Podemos extender el concepto de valor absoluto de los números reales para definir el módulo de un número complejo $z = (x,y)$ como

$$|z| = \sqrt{x^2+y^2}$$

La distancia entre los puntos $z_1 = (x_1,y_1), z_2 = (x_2,y_2)$ está dado por 

$$|z_1-z_2| = \sqrt{(x_1-x_2)^2+(y_1-y_2)^2}$$

En Python para calcular el módulo de un número complejo usamos la función **abs**

````python
abs(4+95j)
````

De aquí podemos observar que 

$$Re(z) \leq |Re(z)|\leq |z|$$

$$Im(z) \leq |Im(z)|\leq |z|$$

Además tenemos la siguiente desigualdad

$$||z_1|-|z_2||\leq |z_1\pm z_2|\leq |z_1|+|z_2|$$

### El conjugado

El conjugado de un número complejo $z = x+iy$ se define como el número $\bar{z} = x-iy$. En el plano complejo, el conjugado de un número complejo $(x,y)$ es su reflexión sobre el eje real $(x,-y)$.

En Python para obtener el conjugado de un número complejo usamos la función **conjugate**

````python
(1+1j).conjugate()
````

Observe que 

$$\bar{\bar{z}} = z$$

Y que 

$$|\bar{z}| = |z|, \forall z\in \mathbb C$$

#### Función **solve_poly_inequality**

La función **solve_poly_inequality** se utiliza para resolver inecuaciones polinomiales.

Primero creamos una variable que contenga la inecuación deseada a resolver.

$$x^3 - x > 0$$

**NOTA** : Las expresiones deben ir completamente del lado izquierdo del operador lógico o completamente del lado derecho.

````python
poly_inequality = (x**3)-x > 0
````

Lo siguiente es extraer la expresión que está al lado izquierdo del operador lógico (en este caso usamos la función **lhs** para dicho propósito, pero si fuera del lado derecho sería **rhs**)

````python
left_side = poly_inequality.lhs
````

Luego creamos un polinomio como lo define Sympy con la función **Poly**, esta función recibe como primer parámetro toda la expresión polinomial de la inecuación a estudiar y como segundo la variable que almacena el símbolo matemático que lo contiene (en este caso la variable 'x')

````python
Pol = sympy.Poly(left_side,x)
````

Ahora necesitamos extraer el operador lógico que creamos en nuestra variable original con la función **rel_op**

````python
Rel = poly_inequality.rel_op
````

Usamos la función **solve_poly_inequality** que recibe como primer argumento la variable que aplicamos la función **Poly** y como segundo argumento la variable que contiene el operador lógico.

````python
sympy.solve_poly_inequality(Pol,Rel)
````

Notemos que el resultado es una lista que contiene los intervalos donde se satisface dicha desigualdad.

#### Función **solve_rational_inequalities**

La función **solve_rational_inequalities** sirve para resolver funciones racionales.

La metodología es análoga a la solución de inecuaciones polinomiales, la principal diferencia es que ahora debemos definir el numerador y denominador.

Resolvamos la siguiente inecuación

$$\frac{x+3}{x-1} > 0, x\neq 1$$

Los primeros dos pasos son los mismos, asignamos a una variable que almacene la desigualdad y luego extraemos la parte izquierda (o derecha) de la expresión

````python
rational_inequality = ((x+3)/(x-1))>0
left_side2 = rational_inequality.lhs
````

Usando la función **as_numer_denom** definidos quien es el numerador y denominador

````python
numerador, denominador = left_side2.as_numer_denom()
````

Usamos la función **Poly** pero ahora aplicado tanto el numerador y denominador anterior definidos

````python
p1 = sympy.Poly(numerador)
p2 = sympy.Poly(denominador)
````

Extraemos el operador lógico con la función **rel_op**

````python
rel = rational_inequality.rel_op
````

Usamos la función **solve_rational_inequalities**, en este caso particular el argumento será una lista que contenga una lista y a su vez una tupla que especifique el numerador, denominador y el operador

````python
sympy.solve_rational_inequalities([[ ( (p1,p2), rel ) ]])
````

#### Función **solve_univariate_inequality**

La función **solve_univariate_inequality** sirve para resolver inecuaciones no polinomiales.

Resolvamos la siguiente inecuación

$$\tan(x) > 0$$

Primero creamos una variable que contenga la inecuación

````python
trig_inequation = sympy.tan(x) > 0
````

Notemos que para la función $\tan(x)$ la importamos de la librería Sympy, esto mismo sucederá para todas las demás funciones trigonométricas, hiperbólicas, etc.

Usamos la función **solve_univariate_inequality**, el primer argumento es la variable que contiene la desigualdad, el segundo argumento es la variable que contiene el símbolo matemático, en este caso 'x'.

````python
sympy.solve_univariate_inequality(trig_inequation,x,relational=False)
````

El tercer argumento 'relational = False' le indica a la función que nos regresa la solución como un conjunto.

Con todo lo anterior, proponemos en este COLAB una función que arroje la solución de cualquier inecuación.

````python
def solve_inequation (exp,var):
  '''
  exp : Recibe una variable que contenga la expresion matematica a estudiar. La expresion matematica
        debe ir totalmente del lado izquierdo, i.e, x**2 + 2 < 0, x+y > 0, (x+2)/(x+4) > 0.

  var : Recibe la variable que contiene el simbolo matematico a estudiar.
  '''
  from sympy import Poly, solve_poly_inequality, solve_rational_inequalities, solve_univariate_inequality
  lhs = exp.lhs
  if lhs.is_polynomial() == True:
    p = Poly(lhs,var)
    relation = exp.rel_op
    return solve_poly_inequality(p,relation)
  elif lhs.is_rational_function() == True:
    numer, denom = lhs.as_numer_denom()
    p1,p2 = Poly(numer), Poly(denom)
    relation = exp.rel_op
    return solve_rational_inequalities([[((p1,p2),relation)]])
  else:
    return solve_univariate_inequality(exp,var,relational = False)
````

## Series

Sea $\{a_n\}$ una sucesión de números reales. Considere la siguiente sucesión que se forma a partir de la sucesión dada

$$S_1 = a_1$$
$$S_2 = a_1 + a_2$$
$$S_3 = a_1 + a_2 + a_3$$
$$\vdots$$
$$S_n = a_1 + \ldots + a_n$$
$$\vdots$$

$S_n$ es conocido como suma parcial. Entonces llamamos serie a la sucesión

$$\{S_1,S_2,\ldots ,S_n,\ldots \}$$

Sí $\{S_n\}$ converge tenemos que

$$\lim_{n\to \infty}S_n = S$$

Decimos entonces que la serie converge y que su suma es $S$, lo cual se denota como

$$\sum_{n=1}^{\infty}a_n = S$$

Supongamos que queremos escribir hasta el $n$ término de la siguiente serie en Python

$$\sum_{i=0}^{n}\frac{x^i}{i!} = 1 + x + \frac{x^2}{2!} + \ldots + \frac{x^n}{n!}$$

Como $n$ es un valor que se debe asignar previamente entonces no es redituable hacer una asignación manual, por ello crearemos una función que vaya acumulando la suma de todos los elementos y que, si lo deseamos, pueda calcular el valor de esta serie

````python
def Serie_1 (n,valor):
  '''
  n : Es un numero natural que indica el numero de elementos de la serie finita que se aproxima
      al valor e
  
  valor : Es un numero real que se sustituirá en la variable en juego. Si valor = None entonces solamente
          la función regresa la serie simbólica
  '''
  from sympy import Symbol, pprint
  from math import factorial
  var, sum = Symbol('x'), 0
  for i in range (n+1):
    sum += (var**i)/factorial(i)
  if valor != None:
    return float(sum.subs({var:valor}))
  else:
    return pprint(sum,order='rev-lex')
````

Imprimimos solamente 10 elementos de la serie que se aproxima al valor $e$, decimos que se aproxima y no es porque necesitariamos infinitos sumandos.

````python
Serie_1(10,None)
````

## Conjunto $\emptyset$

Antes de hablar de $\emptyset$ en Python necesitamos conocer ciertos axiomas.

Hay un conjunto que no tiene elementos

***Axioma*** 2

Si todo elemento de $X$ es un elemento de $Y$, y todo elemento de $Y$ es un elemento de $X$, entonces $X=Y$

$$X=Y \Leftrightarrow (\forall x : x\in X \implies x\in Y)\wedge (\forall x : x \in Y \implies x \in X)$$

**Proposición**

Hay un único conjunto que no tiene elementos

*Demostración*

Sean $A,B$ dos conjuntos distintos entre que no tienen elementos.

Como $B$ no tiene elementos entonces todo elemento de $A$ es un elemento de $B$. Como $A$ no tiene elementos entonces todo elemento de $B$ es un elemento de $A$.

Por el axioma 2 podemos decir que $A=B$, lo cual es una contradicción.

$$\therefore A = B$$

**Definición (Conjunto $\emptyset$)**

## Subconjuntos, superconjuntos y conjunto potencia

**Definición (Subconjunto)**

$A$ es un subconjunto de $B$ si cualquier elemento de $A$ pertenece a $B$. En otras palabras, $A$ es un subconjunto de $B$ si

$$\forall x\in A\implies x\in B$$

Esto, matemáticamente, se denota como $A\subseteq B$

**Definición (Subconjunto propio)**

$A$ es un subconjunto propio de $B$ si cualquier elemento de $A$ pertenece a $B$ pero al menos un elemento de $B$ no pertenece a $A$. Lo denotamos como $A\subset B$.

**Definición (Superconjunto)**

$B$ es un superconjunto de $A$ sí $A$ es un subconjunto de $B$.

**Definición (Superconjunto propio)**

$B$ es un superconjunto propio de $A$ sí $A$ es un subconjunto propio de $B$

**Teorema**

Para cualesquiera conjuntos $A,B,C$ tales que $A\subseteq B$ y $B\subseteq C$ se tiene $A\subseteq C$

*Demostración*

Considere $x\in A$, arbitrario. Por hipótesis $A\subseteq B$ entonces $x\in B$, por definición de subconjunto, y además, por hipótesis $B\subseteq C$ entonces como $x\in B\implies x\in C$, por definición de subconjunto.

Como $x$ es arbitrario entonces $\forall x \in A,\implies x\in C$

$$\therefore A\subseteq C$$

**Teorema**

Sea $A$ un conjunto cualquiera, distinto del vacío, se tiene que $\emptyset\subset A$

*Demostración*

Supongamos que $\emptyset \not \subset A$, esto implica que existe al menos un elemento de $\emptyset$ que no es elemento de $A$ sin embargo esto es contradictorio porque $\emptyset$ no tiene elementos.

$$\therefore \emptyset \subset A$$

**Definición (Conjunto potencia)**

Para cualquier conjunto $X$ existe un conjunto $S$ tal que $A\in S$ si y sólo si $A\subseteq X$. Llamamos a $S$ el conjunto potencia de $X$ y denotamos como $\rho(x)$

**Teorema**

Sí $A\subseteq B$ entonces $\rho(A)\subseteq \rho(B)$

**Teorema**

Sea $A$ un conjunto, entonces $|\rho(A)| = 2^{|A|}$

**Teorema**

Considere el siguiente conjunto

$$\{1,3,5,7,9\}$$

$$\{1,3\}$$

## Unión, intersección y producto cartesiano

**Definición (Intersección)**

Sí $A,B$ son conjuntos, la intersección de $A$ y $B$ es el conjunto 

$$A\cap B = \{x : (x\in A)\wedge (x\in B)\}$$

**Definición (Conjuntos ajenos)**

Diremos que los conjuntos $A$ y $B$ son ajenos sí $A\cap B = \emptyset$

**Definición (Unión)**

Sí $A,B$ son conjuntos, la unión de $A$ y $B$ es el conjunto

$$A\cup B = \{x : (x\in A) \vee (x\in B)\}$$

**Definición (Par ordenado)**

Se define el par ordenado de elementos $a$ y $b$ como

$$(a,b) = \{\{a\},\{a,b\}\}$$

**Definición (Producto cartesiano)**

Sean $A$ y $B$ conjuntos cualesquiera. El producto cartesiano de $A$ y de $B$ es el conjunto $A\times B$ es el conjunto consistente de todos aquellos pares ordenados $(a,b)$ tales que $a\in A, b\in B$, esto es,

$$A\times B = \{(a,b) : a\in A \wedge b \in B\}$$

**Teorema**

Para cualesquiera conjuntos $A,B,C,D$ tenemos 

* $A\cap B \subseteq A \subseteq A\cup B$

* Sí $A\subseteq C$ y $B\subseteq D$ entonces $A\cap B \subseteq C\cap D$ y $A\cup B\subseteq C\cup D$

* $A\subseteq C$ y $B\subseteq C$ si y sólo si $A\cup B \subseteq C$

**Teorema**

La $\cap$ y $\cup$ satisfacen la reflexividad, asociatividad y conmutatividad.

**Teorema**

Los siguientes enunciados son equivalentes 

* $A\subseteq B$

* $A = A\cap B$

* $B = A\cup B$

**Teorema**

$(a,b) = (c,d)$ si y sólo si $a=c, b = d$

**Teorema**

Para cualesquiera $A,B$, conjuntos, $A\times B$ es conjunto.

**Teorema**

$|X\times Y| = |X|\cdot |Y|$

## Probabilidad

**Definición (Experimento)**

El proceso por medio del cual se obtiene una medición

**Definición (Espacio muestral)**

Es un conjunto formado por todos los posibles resultados de un experimento, se denota por $S$.

**Definición (Evento)**

Es el conjunto de los resultados deseados

**Teorema**

Dada una distribución uniforme (donde cada posible resultado tiene la misma probabilidad de éxito) entonces la probabilidad de un evento $P(E)$ está dado de la siguiente manera

$$P(E) = \frac{|E|}{|S|} \in [0,1]$$

Donde $E$ es el evento y $S$ el espacio muestral.

Por ejemplo, si dado el siguiente conjunto $\{1,2,3,4,5,6\}$, ¿cuál es la probabilidad de obtener $1$?.

Nuestro evento está dado como $E = \{1\}$ y $S = \{1,2,3,4,5,6\}$.

$$\therefore P(E) = \frac{|E|}{|S|} = \frac{1}{6}$$

Como vamos a manejar conjuntos finitos y, hasta este punto, los conjuntos a manejar son de distribución uniforme creamos una función llamada 'probabilidad' que nos ayude a calcular la probabilidad de un cierto evento.

````python
def Probabilidad (S,E):
  '''
  S : Es el conjunto muestral

  E : Es el evento
  '''
  return len(E)/len(S)
````

Supongamos que tenemos en una bolsa los números del 1 al 100, queremos saber cual es la probabilidad de sacar un número primo en dicha bolsa.

Primero creamos una función que nos ayude a decir si un número es primo.

````python
def Primo (Numero):
  for It in range (2,Numero):
    if Numero%It == 0:
      return False
  return True
````

Creamos el conjunto 'E' al cual le iremos agregando todos los primos que hay entre 1 y 100.

````python
E = sympy.FiniteSet(*list(Num_Primo for Num_Primo in range (2,101) if Primo(Num_Primo) == True))
````

Usamos la función probabilidad anteriormente definida

````python
Probabilidad(sympy.FiniteSet(*range(1,101)), E)
````

**Teorema**

Sea $-\infty \leq a < b \leq \infty$ y sea $f,g$ diferenciables en $(a,b)$ tal $g^{'}(x)\neq 0, \forall x\in (a,b)$. Suponga que 

$$\lim_{x\to a^{+}}f(x) = \lim_{x\to a^{+}}g(x) = 0$$

* Sí $\lim_{x\to a^+}\frac{f^´(x)}{g^´(x)} = L\in \mathbb R\implies \lim_{x\to a^+}\frac{f(x)}{g(x)} = L$

* Sí $\lim_{x\to a^+}\frac{f^´(x)}{g^´(x)} = L\in \{-\infty,\infty\}\implies \lim_{x\to a^+}\frac{f(x)}{g(x)} = L$

**Teorema**

Sea $-\infty \leq a < b \leq \infty$ y sea $f,g$ diferenciables en $(a,b)$ tal $g^´(x)\neq 0,\forall x\in (a,b)$. Suponga que

$$\lim_{x\to a^+}g(x) = \pm\infty$$

* Sí $\lim_{x\to a^+}\frac{f^´(x)}{g^´(x)} = L\in \mathbb R\implies \lim_{x\to a^+}\frac{f(x)}{g(x)} = L$

* Sí $\lim_{x\to a^+}\frac{f^´(x)}{g^´(x)} = L\in \{-\infty,\infty\}\implies \lim_{x\to a^+}\frac{f(x)}{g(x)} = L$

# Optimización con Python

Si bien existe una librería encargada de los Métodos de Optimización en Python aquí vamos a desarrollar dos métodos simples a modo de práctica. 

[Optimization with Python][1]

[1]: https://docs.scipy.org/doc/scipy/reference/optimize.html

[2]: https://pypi.org/project/PuLP/

Primero iniciaremos con unas definiciones y resultados que nos ayudarán a atacar los problemas planteados.

**Definición (Diferenciabilidad)**

Sea $f(x)$ una función escalar de $n$ dimensiones con dominio $\mathbb R^n$ e imagen $\mathbb R$, decimos que es diferenciable en un punto $x$ sí podemos encontrar un vector $a\in \mathbb R^n$ tal que para toda $y\in \mathbb R^n$ sucede:

$$f(x+y) = f(x) + (a,y) + O(y)$$

El vector $a$ es llamado la derivada o gradiente de $f(x)$ en el punto $x$ y se denota por $f'(x)$ o $\nabla f(x)$. Entonces

$$f(x+y) = f(x) + (\nabla f(x),y) + O(y)$$

$O(y)$ es la función, la cual satisface las condiciones siguientes

1. $O(0) = 0$

2. $\lim_{||y||\to 0}O(y) = 0 \exists$

3. $\lim_{||y||\to 0}\frac{O(y)}{||y||} = 0$

**Teorema**

Si $f(x)$ es diferenciable en $x$ entonces es continua en $x$.

**Teorema**

Sí $f(x)$ es diferenciable en $x$ entonces $\nabla f(x)$ es único y es de la forma $\nabla f(x) = \left(\frac{\partial f}{\partial x_1}(x),\ldots, \frac{\partial f}{\partial x_n}(x)\right)$

**Definición (Mínimo local)**

Sea $x^*\in \mathbb R^n$, decimos que $x^*\in \mathbb R^n$ es un punto mínimo local sí $\exists \varepsilon > 0$ tal que $\forall x \in ||x-x^*||<\varepsilon$ se cumple

$$f(x^*)\leq f(x)$$

**Definición (Mínimo global)**

Sea $x^*\in \mathbb R^n$, diremos que $x\in \mathbb R^n$ es punto mínimo global sí $\forall x\in \mathbb R$ se cumple

$$f(x^*)\leq f(x)$$

**Teorema (Fermat)** «Condición necesaria de la condición de primer orden»

Sea $x^*\in \mathbb R^n$ tal que $x^*$ es un punto mínimo local de $f(x)$ en $\mathbb R^n$ y sea $f(x)$ diferenciable en $x^*$. Entonces

$$\nabla f(x^*) = 0$$

**Teorema** «Condición necesaria de la condición de segundo orden»

Sea $x^*$ un punto mínimo local de $f(x)$ en $\mathbb R^n$ y sea $f(x)$ dos veces diferenciable en el punto $x^*$.

$$\implies \nabla^2 f(x) \geq 0$$

**Teorema** «Condición suficiente de la función de segundo orden»

En el punto $x^*$ sea $f(x^*)$ dos veces diferenciable, además considere $\nabla f(x^*) = 0$ y $\nabla^2 f(x^*)> 0$

$$\implies x^*\text{ mínimo local}$$

**Teorema (Weierstrass)** 

Sea $f(x)$ una función continua en $\mathbb R^n$ y sea $Q_{\alpha} = \{x : f(x) \leq \alpha \}$, para algún $\alpha$, donde $Q_{\alpha}$ es distinto del vacío y acotado.

Entonces existe un punto mínimo global de $f(x)$ en $R^n$.

**Definición (Función Convexa)**

Sea $f(x)$ una función escalar en $\mathbb R^n$. Decimos que $f(x)$ es convexa sí $\forall x,y \in \mathbb R^n, 0\leq \lambda \leq 1$ sucede que

$$f(\lambda x + (1-\lambda)y)\leq \lambda f(x) + (1-\lambda)f(y)$$

**Teorema**

Sea $\rho (t)$ una función diferenciable en $\mathbb R$. La convexidad de la función $\rho (\tau)$ es equivalente a la monotonía de la derivada, i.e,

$$\rho' (\tau_1)\geq \rho' (\tau_2),\forall \tau_1\geq \tau_2$$

**Teorema**

Para una función diferenciable $f(x)$ en $\mathbb R^n$, la convexidad es equivalente a la inecuación

$$f(x+y) \geq f(x) + (\nabla f(x),y)$$

**Teorema**

Sea $f$ diferenciable en $\mathbb R^n$

$$f(x+y) \geq f(x) + (\nabla f(x),y) \Leftrightarrow (\nabla f(x)- \nabla f(y),x-y)\geq 0$$

**NOTA** : El teorema anterior nos indica que una función es convexa si y sólo si es monótomanete creciente.

**Teorema**

Una función es convexa si y sólo si $\nabla^2 f(x) \geq 0$

**Teorema**

Sea $f(x)$ convexa y diferenciable en el punto $x^*$ y $\nabla f(x^*) = 0$. Entonces $x^*$ es punto mínimo global de $f(x)$ en $\mathbb R^n$.

## Problemas de Optimización sin restricciones

En este apartado nos enfocaremos a resolver problemas de optimización sin restriccciones para la función objetivo. En general nuestro problema es :

Sea $f : \mathbb R^n \to \mathbb R$ diferenciable.
$$f(x)\to \text{extr}$$

Vamos a resolver un problema de manera manual para después proponer una función en Python que ejecute dicho problema.

**EJEMPLO**

Encontrar los extremos de

$$f(x,y) = 3x^2+4xy+y^2-8x-12y$$

*Solución*

Primero calculamos el gradiente de la función.

$$\nabla f(x,y) = (6x+4y-8,4x+2y-12)$$

Igualamos a $0$ el gradiente y encontramos los puntos que satisfacen dicha condición.

$$\nabla f(x,y) = 0 \Longleftrightarrow 6x+4y-8 = 0,4x+2y-12 = 0$$

Resolviendo el sistema dado tenemos que

$$\nabla f(x,y) = 0 \Longleftrightarrow (x,y) = (8,-10)$$

Calculamos $\nabla^2 f(x,y)$

\begin{equation}
  \nabla^2 f(x,y) = \begin{pmatrix}
  \frac{\partial^2 f}{\partial x^2} & \frac{\partial^2 f}{\partial x\partial y}\\
  \frac{\partial^2 f}{\partial y\partial x} & \frac{\partial^2 f}{\partial y^2}
  \end{pmatrix}
\end{equation}

\begin{equation}
  \nabla^2 f(8,-10) = \begin{pmatrix}
  6 & 4\\
  4 & 2
  \end{pmatrix}
\end{equation}

Notemos que $\nabla^2 f(x,y)$ no es definida positiva pues aunque $6>0$ tenemos que

\begin{equation}
\begin{vmatrix}
6 & 4\\
4 & 2
\end{vmatrix}=-4<0
\end{equation}

Por lo cual el punto encontrado es punto estacionario de $f(x)$, es decir, $f(x)$ no tiene ni mínimo local ni mínimo global. 

Como nos piden los extremos entonces analicemos sí $f(x,y)$ tiene máximo local o máximo global mediante $-f(x,y)$.

Notemos que para $-f(x,y)$ el punto $(8,-10)$ satisface que $-\nabla f(x,y) = 0$. Además note que

\begin{equation}
  -\nabla^2 f(8,-10) = \begin{pmatrix}
  -6 & -4\\
  -4 & -2
  \end{pmatrix}
\end{equation}

Notemos que $-6<0$ y por tanto $-\nabla^2 f(x,y)$ no tiene ni máximo local ni máximo global pues su matriz hessiana no es definida positiva y $-f(x,y)$ no es convexa.

$$\therefore f\text{ no tiene ni mínimo local, global ni máximo local global.}$$

Se propone la siguiente función que analice si la función tiene mínimo o máximo local o global.

````python
def Optimizacion_No_Restricciones (function,variables,function_convex,obj):
  '''
  function : La función objetivo escrito en formato Sympy.

  variables : Una lista que contenga las variables en juego, estas deben estar
              asignadas como 'real = True'.

  function_convex : Recibe un valor booleano. Si es True entonces la funcion
                    es convexa, False en caso contrario.

  obj : Recibe 'min' si deseamos minimizar la funcion objetivo, 'max' en caso
        de maximizar la funcion objetivo.
  '''
  def def_pos(A):
      import numpy as np
      M = np.matrix(A)
      return np.all(np.linalg.eigvals(M+M.transpose()) > 0)
  from sympy import Derivative, solve
  import pandas as pd
  import numpy as np
  if obj == 'min': #CASO min
    Gradiente = list(Derivative(function,variable).doit() for variable in variables)
    T_F = solve(Gradiente, dict = True)
    if function_convex == True: #SI f ES CONVEXA
      return T_F #REGRESAMOS LOS PUNTOS GLOBALES
    else: #SI f NO ES CONVEXA
      Res = []
      for Punto in T_F: #CREAMOS LA MATRIZ HESSIANA PARA EL PUNTO K
        H = pd.DataFrame(np.zeros( ( len(variables),len(variables)) ) )
        for Fila in range (len(variables)): #TOMAMOS LA FILA i Y LE AGREGAMOS LAS SEGUNDAS DERIVADAS PARCIALES EVALUADAS EN EL PUNTO K 
          H.iloc[Fila] = list( Derivative(Derivada,variables[Fila]).doit().subs(Punto) for Derivada in Gradiente)
        H = np.array(H, dtype=float)
        if def_pos(H) == True:
            Res.append(Punto)
      return Res
  elif obj == 'max': #CASO max
    function = -function
    Gradiente = list(Derivative(function,variable).doit() for variable in variables)
    T_F = solve(Gradiente, dict = True)
    if function_convex == True: #SI f ES CONVEXA
      return T_F #REGRESAMOS LOS PUNTOS GLOBALES
    else: #SI f NO ES CONVEXA
      Res = []
      for Punto in T_F: #CREAMOS LA MATRIZ HESSIANA PARA EL PUNTO K
        H = pd.DataFrame(np.zeros((len(variables),len(variables))))
        for Fila in range (len(variables)): #TOMAMOS LA FILA i Y LE AGREGAMOS LAS DERIVADAS PARCIALES EVALUADAS EN EL PUNTO K
          H.iloc[Fila] = list( Derivative(Derivada,variables[Fila]).doit().subs(Punto) for Derivada in Gradiente)
        H = np.array(H, dtype=float)
        if def_pos(np.array(H)) == True:
          Res.append(Punto)  
      return Res
  else:
      while obj != 'min' and obj != 'max':
          obj = input('Ingresa (min/max) : ')
````

La función anterior necesita, por ahora, que de manera manual se determine si la función a analizar es convexa. También se añade que las variables independientes deben ser definidas en $\mathbb R$

Veamos unos ejemplos para ver el funcionamiento de la función propuesto anteriormente.

Calcule los máximos y mínimos, si existen, de

$$f(x,y,z) = x^2 + y^2 + z^2 - xy + x -2z$$

Primero veamos si la función en cuestión es convexa.

\begin{equation}
  \nabla^2 f(x,y,z) = \begin{pmatrix}
  2 & -1 & 0\\
  -1 & 2 & 0\\
  0 & 0 & 2  \end{pmatrix} > 0, \forall x,y,z\in \mathbb R
\end{equation}

$$\therefore f\text{ es convexa}$$

Definimos $x,y,z\in \mathbb R$.

````python
x,y,z = sympy.Symbol('x',real = True),sympy.Symbol('y', real = True),sympy.Symbol('z', real = True)
````

Veamos si tiene mínimos globales (pues $f$ es convexa).

````python
Optimizacion_No_Restricciones(function = (x**2)+(y**2)+(z**2)-x*y+x-2*z,
                              variables = [x,y,z],
                              function_convex = True,
                              obj = 'min')
````

$$\therefore \left(-\frac{2}{3},-\frac{1}{3},1\right)\text{ es mínimo global de }f$$

Calcule los mínimos y máximos, si existen, de

$$g(x,y) = 3xy - x^2y -xy^2$$

Notemos que $g$ no es convexa pues 

\begin{equation}
  \nabla^2 f(x,y) = \begin{pmatrix}
  -2y & 3-2x-2y\\
  3-2x-2y & -2x
  \end{pmatrix}
\end{equation}

No satisface que sea definida positiva para todo $x,y\in \mathbb R$.

Busquemos sus mínimos locales, si existen.

````python
Optimizacion_No_Restricciones(function = 3*x*y-(x**2)*y-x*(y**2),
                              variables = [x,y],
                              function_convex = False,
                              obj = 'min')
````

$$\therefore f\text{ no tiene mínimos locales}$$

Busquemos sus máximos locales, si existen.

````python
Optimizacion_No_Restricciones(function = 3*x*y-(x**2)*y-x*(y**2),
                              variables = [x,y],
                              function_convex = False,
                              obj = 'max')
````

$$\therefore f\text{ tiene un máximo local en }(1,1)$$

Veamos la gráfica en tres dimensiones de la función anterior.
